```python
'''
    20153237 김현준
    7월 15일 수업 소스
'''

import urllib
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import re

url = input('url을 입력하시오 : ')

# fh = urllib.request.urlopen('https://www.google.com')
fh = urllib.request.urlopen(url)

data = ''
count = 0

for line in fh:
    #print(line.decode().strip()) # binary 값을 아스키 값으로 바꾸기 위해 decode 를 한다.
    line = line.decode().strip()
    
    for value in line:
        
        data += value
        count += 1


bodyData = re.findall('<body.*<\/body>', data)



print(bodyData)
print(count)
```

```python
needData = re.findall('<[^>]*>([^<]*)<[^>]*>', bodyData[0])

for value in needData:
    value = value.replace(' ', '')
    
    if value != '':
        print(value)
'''
data = 전체 데이터를 바디만 뽑은거

t1 = re.sub(re.compile('<.*?>'), '', data)
-> 바디만 뽑은 데이터에서 <> 를 포함한 안에 들어간 값을 찾아 sub 함수로 삭제한다.
t2 = bs(data, 'html.parser')

'''
```

```python
# https://poloniex.com/public?command=returnChartData
# &currencyPair=USDT_BTC # 변경됨 
# &start=1546300800&end=9999999999
# &period=14400
# 차트 데이터를 준 사이트 

import time
import requests
import pandas as pd

# poloniex에 있는 일일 데이터를 얻기 위해 request 한다.
poloniex_daily_data = requests.get(
        'https://poloniex.com/public?command=returnChartData&'+
        'currencyPair=USDT_BTC&' +
        'start='+
        str(int(time.mktime((2019,1,1,0,0,0,0,0,0)))) + # 보고싶은 시간을 정한다.
        '&end=9999999999&period=86400' # 가장 최근까지 보고싶다면 9999999999
)

btc_df = pd.DataFrame(poloniex_daily_data.json()) # 제이슨 형식으로 바꾸어서 데이터프레임으로 만들면 이쁘게 만들어진다.
```

```python
def btc_price(yyyy, mm, dd):
    poloniex_daily_data = requests.get(
        'https://poloniex.com/public?command=returnChartData&'+
        'currencyPair=USDT_BTC&' +
        'start='+
        str(int(time.mktime((yyyy,
                             mm,
                             dd,
                             0,0,0,0,0,0)))) + # 보고싶은 시간을 정한다.
        '&end=9999999999&period=86400' # 가장 최근까지 보고싶다면 9999999999
    )
    
    return pd.DataFrame(poloniex_daily_data.json()) 
 
yyyy, mm, dd = map(int, input("yyyy mm dd 입력").split())
```

```python
test = btc_price(yyyy, mm, dd)
```

```python
from xml.etree import ElementTree as et

data = '''
<person>
    <name>Chuck</name>
    <phone type="intl">
        +1 734 303 4456
    </phone>
    <email hide="yes"/>
</person>'''

tree = et.fromstring(data)
print('Name:',tree.find('name').text)
print('Attr:',tree.find('email').get('hide'))
```

```python
input = '''
<stuff>
    <users>
        <user x="2">
            <id>001</id>
            <name>Chuck</name>
        </user>
        <user x="7">
            <id>009</id>
            <name>Brent</name>
        </user>
    </users>
</stuff>'''

stuff = et.fromstring(input)
#input 데이터에 대해 데이터처리가 시작됨을 설정한다.

lst = stuff.findall('users/user')
# users 안에 존재하하는 user에 대한 정보들을 가지고 있다.

print('User count:', len(lst))
# users태그 안 user의 수가 몇인지 알 수 있따.

for item in lst:
    
    print('Name', item.find('name').text)
    print('Id', item.find('id').text)
    print('Attribute', item.get('x'))
```

```python
# jason 자료 구조는 거의 모든 언어가 이와 같은 형식의 자료형이 있기 때문에 
# 자료의 교한에 상당히 유리하다. 형식도 간편하기 때문에 자주 쓰인다.

import json

input = '''
[
    { 
        "id" : "001",
        "x" : "2",
        "name" : "Chuck"
    } ,
    { 
        "id" : "009",
        "x" : "7",
        "name" : "Chuck"
    }
]
'''
info = json.loads(input)
print ('User count:', len(info))

for item in info:
    print ('Name', item['name'])
    print ('Id', item['id'])
    print ('Attribute', item['x'])
```
